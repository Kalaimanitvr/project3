```python
'''
Task 1
The advertising budget spent on promoting premium computers
in 2023 is increased as compared to 2022. The mean
advertising budget was 221.3 billion dollars in 2022 and it was
222.2 billion dollars in 2023. A promoter in this company
believes that the average advertising budget is higher than that
of 2022. Priya, a data analyst, randomly selected 40 premium
computers to check this notion.
Use a 5% level of significance to test Maya's hypothesis. Consider normally distribution in the population and standard deviation is 74.83.

'''
print("Null Hypothesis : The mean advertising budget in 2023 is equal to or less than the mean in 2022.\nAlternative Hypothesis : The mean advertising budget in 2023 is greater than the mean in 2022.")

import statistics as st
import math
from scipy.stats import norm

sm = 222.2  
pm = 221.3  
psd = 74.83  
n= 40 
alpha=.05

print(f'sample_mean:{sm}')
print(f"alpha:{alpha}")
print(f'pop_mean:{pm}')
print(f"pop_std:{psd}")
print(f'Sample_size:{n}')
z_critical= norm.ppf(1-(alpha))
print(f'z critical:{z_critical}')

z_stats=(sm-pm)/(psd/math.sqrt(n))
print(f'Z_stats:{z_stats}')

p_val=norm.sf(abs(z_stats))
print(f'P_val:{abs(p_val)}')

if p_val<alpha:
    print('reject the null')
else:
    print('accept the null')
print('-----------------------------------------------------------------------------------------------------------------------------------------')
print("Since the p-value (0.470) is greater than the level of significance is alpha=0.05, we fail to reject the null hypothesis. This means there isn't enough evidence to support the claim that the average advertising budget in 2023 is higher than that of 2022.")
```

```python
'''
Task 2
Is there a statistically significant difference 
in the average price of computers with CD players and computers without CD players? Use 5% of the significance level for the test.

'''
print("Null Hypothesis: There is no significant difference in the average price of computers with CD players and computers without CD players\nAlternative Hypothesis : There is a significant difference in the average price of computers with CD players and computers without CD players.")

import pandas as pd
import numpy as np
from scipy.stats import t
data=pd.read_csv("DS1_C8_Computers_Data_Project (1).csv")
cd_with =data[data['cd']=='yes']
cd_without =data[data['cd']=='no']
sm_cd=cd_with['price'].mean()
sm_cdwithout=cd_without['price'].mean()
ssd_cd=cd_with['price'].std()
ssd_cdwithout=cd_without['price'].std()

var_cd=ssd_cd**2
var_cdwithout=ssd_cdwithout**2

dof=len(cd_with)+len('cd_without')-2
print(f"sample mean price for with cd players: {sm_cd}")
print(f"sample mean price for without cd players: {sm_cdwithout}")
print(f"sample sd price for with cd players: {ssd_cd}")
print(f"sample sd price for without cd players: {ssd_cdwithout}")
alpha=.05


num=abs(sm_cd- sm_cdwithout)
denom= math.sqrt(var_cd*(len(cd_with)-1)+ var_cdwithout*(len('cd_without')-1)) *math.sqrt((1/len('cd_with')) + (1/len('cd_without')))
tstats=num/denom
print("T Statistics: ", tstats) 
t_critical=t.ppf(alpha/2, dof)
print("T Critical: ",t_critical)
p_val= t.sf(abs(tstats), dof)*2
print("P value: ",p_val)
if p_val< 0.05:
    print("Null Hypothesis is rejected There is no difference between means of two groups")
else:
    print("Null hypothesis is accepted There is a difference between means of two groups")

print('\n\ninterpretation:\nBased on the t-test results, the p-value (0.9869) is greater than the significance level (0.05), and the t-statistic (0.0164) is within the critical range, which means the null hypothesis is accepted. Thus, there is no statistically significant difference in the average prices of computers with and without CD players, and any observed difference is likely due to random variation.')
```

```python
'''
task 3
b) Analyze price disparity between premium and non-premium computers.
Is there a significant difference in the mean prices of premium and non-premium computers?
Assume that the prices are normally distributed and that the population variances are approximately equal.
Use a 5% significance level to test this hypothesis.
'''
print("Null Hypothesis : The mean price of premium computers is equal to the mean price of non-premium computers.\nAlternative Hypothesis : The mean price of premium computers is different from the mean price of non-premium computers.")

from scipy.stats import ttest_ind
prices_premium = data[data['premium'] == 'yes']['price']
prices_non_premium = data[data['premium'] == 'no']['price']


t_stat, p_value = ttest_ind(prices_premium, prices_non_premium, equal_var=True)


significance_level = 0.05


# Output the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
print('-----------------------------------------------------------------------------------------------------------------------------------------')
if p_val<0.05:
    print('Null Hypothesis is rejected')
else:
    print('Null Hypothesis is accepted ')
print('-----------------------------------------------------------------------------------------------------------------------------------------')
print("If the null hypothesis is accepted, it means that there is no significant difference between the mean prices of premium and non-premium computers, implying that their average prices are statistically similar.")
```
